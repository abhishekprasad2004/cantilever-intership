```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Step 1: Define URL and Headers
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
product_list = []
```

```python
# Step 2: Loop through pages
for page in range(1, 51):  # scrape 3 pages as example
    url = base_url.format(page)
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        books = soup.find_all('article', class_='product_pod')

        for book in books:
            title = book.h3.a['title']
            price = book.find('p', class_='price_color').text.strip()
            rating = book.p['class'][1]  # second class is rating
            availability = book.find('p', class_='instock availability').text.strip()

            product_list.append({
                'Title': title,
                'Price': price,
                'Rating': rating,
                'Availability': availability
            })
    else:
        print(f"Request failed at page {page} with status {response.status_code}")


```

```python
# Step 3: Display Data in Colab
df = pd.DataFrame(product_list)
df.head()
```

```python
# Step 4: Save to CSV
df.to_csv('scraped_books.csv', index=False)
print("Saved to scraped_books.csv")
```
